Ce notebook contient un code permettant d'interagir avec l'API de **Edamam Food and Grocery Database** afin d'extraire tous les produits contenant du champagne.

Plus précisément, pour chaque produit, on veut récupérer les données des colonnes `foodId`, `label`, `category`, `foodContentLabel` et `image`.

Le code est divisé en quatre cellules :
- Importations en entêtes
- Requête à l'API
- Construction du dataframe
- Enregistrement en csv

```python
import pandas as pd
import requests

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

headers = {
    "X-RapidAPI-Key": "646c1bc8damsh692dece350ad079p17380fjsn5f25b519072b",
    "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}
```

```python
querystring = {"ingr": "champagne"}
response = requests.get(url, headers = headers, params = querystring)

if response.status_code == 200 :
    data = response.json()
else :
    print("Erreur lors de la récupération des données de l'API.")
```

```python
products = []

for hint in data.get('hints', [])[:10] :
    product = {
        'foodId': hint['food']['foodId'],
        'label': hint['food']['label'],
        'category': hint['food']['category'],
        'foodContentsLabel': hint['food'].get('foodContentsLabel', ''),
        'image': hint['food'].get('image', '')
    }
    products.append(product)

df = pd.DataFrame(products)
df.head(5)
```

```python
df.to_csv('champagne_products.csv', index = False)
print("Les 10 premiers produits à base de champagne ont été enregistrés dans le fichier 'champagne_products.csv'.")
```
